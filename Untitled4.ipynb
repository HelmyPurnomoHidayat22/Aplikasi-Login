```python
# Import library yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
import numpy as np

# Membaca dataset
file_path = 'student_spending_cleaned.csv'
data = pd.read_csv(file_path)

# Memilih kolom finansial sebagai fitur
financial_columns = [
    'monthly_income_idr', 'financial_aid_idr', 'tuition_idr', 'housing_idr',
    'food_idr', 'transportation_idr', 'books_supplies_idr', 'entertainment_idr',
    'personal_care_idr', 'technology_idr', 'health_wellness_idr', 'miscellaneous_idr'
]
financial_data = data[financial_columns]

# Tambahkan fitur baru yang lebih informatif
financial_data = financial_data.copy()
financial_data['expense_to_income_ratio'] = financial_data.sum(axis=1) / data['monthly_income_idr']
financial_data['entertainment_ratio'] = data['entertainment_idr'] / data['monthly_income_idr']
financial_data['essential_expenses'] = (data['food_idr'] + data['housing_idr'] + data['transportation_idr']) / data['monthly_income_idr']
financial_data['education_expenses'] = (data['tuition_idr'] + data['books_supplies_idr']) / data['monthly_income_idr']

# Membuat label yang lebih realistis
data['addiction_level'] = (financial_data['entertainment_ratio'] > 0.2).astype(int)

# Normalisasi data menggunakan RobustScaler
scaler = RobustScaler()
X = scaler.fit_transform(financial_data)

# Target (label kecanduan)
y = data['addiction_level']

# Oversampling dengan SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Membagi dataset dengan stratifikasi
X_train, X_test, y_train, y_test = train_test_split(
    X_smote, y_smote, test_size=0.2, random_state=42, stratify=y_smote
)

# Konversi labels ke numpy array
y_train = np.array(y_train)
y_test = np.array(y_test)

# Hitung class weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = dict(zip(np.unique(y_train), class_weights))

# Model yang lebih kompleks
model = models.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.BatchNormalization(),
    layers.Dropout(0.2),

    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.BatchNormalization(),
    layers.Dropout(0.1),

    layers.Dense(2, activation='softmax')  # Klasifikasi 2 kelas (0 atau 1)
])

# Optimizer dengan learning rate yang lebih rendah
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)

# Kompilasi model
model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks = [
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
]

# Melatih model
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    class_weight=class_weights_dict,
    verbose=1
)

# Evaluasi model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"\nAkurasi Final: {test_accuracy:.4f}")

# Visualisasi hasil training
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
